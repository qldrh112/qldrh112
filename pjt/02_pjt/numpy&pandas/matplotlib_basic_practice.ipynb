```python
# matplotlib 기본 사용법(라인 그래프 그리기)

import matplotlib.pyplot as plt

# 데이터 생성
x = [1, 2, 3, 4, 5]  # x 좌표값
y = [2, 4, 6, 8, 10]  # y 좌표값

# 그래프 그리기
plt.plot(x, y)

# 그래프에 제목과 축 레이블 추가
plt.title('My First Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# 그래프 표시
plt.show()
```

```python
# 두 가지 선 그리기

# 데이터 생성
x = [1, 2, 3, 4, 5]         # x 좌표값
y = [2, 4, 6, 8, 10]        # y 좌표 값
y2 = [3, 6, 9, 12, 15]      # y2 좌표 값

# 그래프 그리기
plt.plot(x, y, label='y')
plt.plot(x, y2, label='y2')     # label = ??

# 범례 추가
plt.legend()

# 그래프에 제목과 축 레이블 추가
plt.title('My Second plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')        # 한글 지원이 안 되나보다. 폰트깨진다.

# 그래프 표시
plt.show()
```

```python
# 막대 그래프 그리기
# 범주형 데이터를 시각화하기 위해 사용, 주로 범주 간의 비교를 위해 사용
# 각 막대는 독립적이며, 막대 간의 간격 존재
# ex) 특정 업종 기업 간 매출액 비교

# 데이터 생성
x = ['Korean', 'Math', 'English', 'Science', 'PE']      # 카테고리
y = [3, 7, 2, 5, 9]                           # 값

# 그래프 그리기
plt.bar(x, y)

# 그래프에 제목과 축 레이블 추가
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')

# 그래프 표시
plt.show()
```

```python
# 히스토그램 그리기
# 연속된 데이터의 분포를 시각화하기 위해 사용됨
# 데이터를 구간으로 나누고, 각 구간에 속하는 데이터의 빈도수 또는 밀도를 나타냄
# 연령 별 인구 분포, 상품의 가격 분포 등

import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
data = np.random.randn(1000) # 랜덤한 데이터

# 그래프 그리기
plt.hist(data, bins=30)

# 그래프에 제목과 축 레이블 추가
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 그래프 표시
plt.show()
```

```python
# 데이터와 함께 활용하기
# csv 파일을 읽어와 Pandas DataFrame 생성하기
import pandas as pd

# Numpy를 활용한 csv file open
def file_open_by_numpy():
    # np.loadtxt(구분자 = ',', 데이터 타입: string)
    np_arr = np.loadtxt('data/test_data.CSV', delimiter=",", encoding='cp949', dtype=str)
    return np_arr

arr = file_open_by_numpy()

# 첫 번째 열을 헤더로 하는 데이터 프레임 생성
header = arr[0, :]              # 첫 번째 행을 헤더로 사용
data = arr[1:, :]               # 첫번째 행을 제외한 나머지 데이터

df = pd.DataFrame(data, columns=header)
print(df.head())
```

```python
print(df.dtypes)
```

```python
# 나이 필드를 숫자 형식으로 변환
df['나이'] = df['나이'].astype(int)

# 나이에 대한 히스토그램 출력
df['나이'].plot(kind='hist', bins=10)

# 그래프 제목 설정
plt.title('Age Distribution')

# x축 레이블 설정
plt.xlabel('Age')

# y축 레이블 설정
plt.ylabel('Count')

# 그래프 출력
plt.show()
```
